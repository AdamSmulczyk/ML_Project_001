 ### <a id="I.1"></a><p style="background-color: #294B8E;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.1 Introduction</p>

The sinking of the **<mark style="background-color:#3b3745;color:white;border-radius:5px;opacity:1.0">Titanic</mark>** is one of the most infamous shipwrecks in history. On April 15, 1912, during her maiden voyage, the widely considered ‚Äúunsinkable‚Äù RMS Titanic sank after colliding with an iceberg. 

Unfortunately, there weren‚Äôt enough lifeboats for everyone onboard, resulting in the death of **<mark style="background-color:#3b3745;color:white;border-radius:5px;opacity:1.0">1502 out of 2224</mark>** passengers and crew. It took about $7.5 million to build the Titanic and it sunk under the ocean due to collision.

This is a very unforgetable disaster that no one in the world can forget. While there was some element of luck involved in surviving, it seems some groups of people were more likely to survive than others.

In this task is to build a predictive model that answers the question: ‚Äúwhat sorts of people were more likely to survive?‚Äù 

 ### <a id="I.2"><p style="background-color: #294B8E;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.2 Dataset Description</p>

Two data sets are provided, a training set and a test set.  The PassengerID variable is assumed to be a random, unique identifier that has no effect on the outcome variable. Therefore, it will be excluded from the analysis. All other variables are potential predictors or independent variables.  The training set therefore consists of 10 predictor attributes, which include:

* <code>PassengerID:</code> Passenger ID
* <code>Survival:</code>  The 'Survival' variable is our outcome or dependent variable. It is a binary nominal data type, where 1 is survived and 0 is not survived.


* <code>Pclass:</code>  The passenger's ticket class. The 'Pclass' variable is an ordinal data type for the 'Ticket' class, a proxy for socio-economic status (SES), representing 1-first (upper class), 2-second (middle class), and 3-third (lower class).


* <code>Name:</code>   The passenger's first and last name. The Name variable is a nominal data type. It can be used in feature engineering to derive gender from title, family size from last name, and SES from titles such as Doctor or Lord. Since these variables already exist, we will use them to see if a title such as Doctor or Lord makes a difference


* <code>Sex:</code>   The passenger's gender. The Sex variables are nominal/categorical data types. For model calculation purposes these will be converted to numeric variables.


* <code>Sibsp:</code>  represents the number of related siblings/spouses on board


* <code>Parch:</code>  represents the number of related parents/children on board. This can be used for feature engineering to create family size.

* <code>Ticket:</code>  ticket number. A random variable at first glance, a unique identifier that has no effect on the outcome variable. The Pclass variable is an ordinal data type for this attribute.


* <code>Fare:</code>  the fare for the ticket.


* <code>Cabin:</code>  cabin, the Cabin variable is a nominal data type that can be used in feature engineering to approximate the position on the ship at the time of the incident and the SES from deck levels. However, because there are many null values, it does not add value and will therefore likely be excluded from the analysis.


* <code>Embarked:</code>  starting port (C = Cherbourg; Q = Queenstown; S = Southampton).The Embarked variables are nominal/categorical data types. For the purposes of calculating the model, they will be converted to numeric variables.

The selected features will allow us to determine the value of the result variable. The value we will be examining/predicting is the attribute "Survival" (0/1).

 ### <a id="I.3.3"><p style="background-color: #294B8E;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">I.3.3 Success metric</p>

The indicators of success specified in the requirements are in this case: **<mark style="background-color:#3b3745;color:white;border-radius:5px;opacity:1.0">Accuracy, F1 score, AUC</mark>** .
Accuracy - this metric tells us about the ratio of well-made classifications to the number of all classifications made. It is expressed in percentages and everyone will understand it immediately. But it also has disadvantages and one of them is a situation in which we have an unbalanced set. Accuracy can therefore be very misleading.

Therefore, an additional indicator of success is the F1 metric, which is the harmonic mean of the two values ‚Äã‚Äãbelow:
* Precision - This metric tells us roughly how much we can trust our positive predictions in a given class. It is defined as:
tp ‚Äì true positive is an observation correctly assigned to the analyzed class
fp ‚Äì false positive is an observation incorrectly assigned.
* Recall - the concept behind recall is similar to precision, but it tells us how many observations we have lost for a given class. Here we use the following equation:

So we can have a situation where accuracy is supposedly quite good, but precision and recall are not so good. That is why it is also useful to consider the F1 metric, which gives us a single number (between 1 and 0, where 1 is the best situation) that evenly considers precision and recall.

 ### <a id="II.2.2"><p style="background-color: #294B8E;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">II.2.2 Statistical Summary üî¢</p>

```python
def statistical_summary_numerical(df: pd.DataFrame):
    df = df.select_dtypes(include='number')
    desc = pd.DataFrame(index=df.columns.to_list())
    desc['type'] = df.dtypes
    desc['count'] = df.count()
    desc['nunique'] = df.nunique()
    desc['%unique'] = desc['nunique'] / len(df) * 100
    desc['null'] = df.isnull().sum()
    desc['%null'] = desc['null'] / len(df) * 100
    desc = pd.concat([desc, df.describe().T.drop('count', axis=1)], axis=1)
    desc["skew"] = df.skew()
    desc["kurtosis"] = df.kurtosis()
    desc = desc.round(2)
    return desc

statistical_summary_numerical(df)
```

```python
def statistical_summary_categorical(df: pd.DataFrame):
    df = df.select_dtypes(include='object')
    desc = pd.DataFrame(index=df.columns.to_list())
    desc['type'] = df.dtypes
    desc['count'] = df.count()
    desc['nunique'] = df.nunique()
    desc['%unique'] = desc['nunique'] / len(df) * 100
    desc['null'] = df.isnull().sum()
    desc['%null'] = desc['null'] / len(df) * 100
    return desc

statistical_summary_categorical(df)
```

```python
numerical_cols = df.select_dtypes(exclude=['object', 'category'])
categorical_cols = df.select_dtypes(include=['object', 'category'])

def fill_empty_categorical(df):
    for i in categorical_cols:
        df[i].fillna(df[i].mode()[0], inplace=True)
    return df

def fill_empty_numerical(df):
    for i in numerical_cols:
        df[i].fillna(df[i].median(), inplace=True)  
    return df

df = fill_empty_categorical(df)
df = fill_empty_numerical(df)


def convert_types (df):
    for i in categorical_cols:
         df[i] = df[i].astype('category')
    for i in numerical_cols:
         df[i] = df[i].astype('int32')  
    return df
df = convert_types(df)
```

```python
# defining age-divided groups
def age_group(age):
    if age < 18:
        return '0-17'
    elif age < 30:
        return '18-29'
    elif age < 50:
        return '30-49'
    else:
        return '50+'

# Group data by age group and passenger class and calculate survival rate
df['AgeGroup'] = df['Age'].apply(age_group)
survival_rates = df.groupby(['AgeGroup', 'Pclass'])['Survived'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.catplot(x='Pclass', y='Survived', hue='AgeGroup', kind='bar', data=survival_rates, palette='viridis')
plt.title('Survival rate by age group and passenger class', fontsize=15)
plt.xlabel('Passenger class', fontsize=12)
plt.ylabel('Survival rate', fontsize=12)
plt.show()
```

 ### <a id="III.4"><p style="background-color: #294B8E;color:white;display: inline-block;padding:.6rem;border-radius:.5rem">III.4 Detect Outliers üßê</p>

```python
def calculate_outliers_percentage(df):

    outlier_counts = {}
    for column in df.select_dtypes(include=['number']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Calculate outliers
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_counts[column] = len(outliers)

    # Print the percentage of outliers for each column
    for column in outlier_counts:
        percentage = (outlier_counts[column] / len(df)) * 100
        print(f"Percentage of outliers in {column}: {percentage:.2f}%")


calculate_outliers_percentage(X)
```
